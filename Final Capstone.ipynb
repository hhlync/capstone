```python
#table from Wikipedia 
url = 'https://en.wikipedia.org/wiki/List_of_districts_of_Seoul'
class_='wikitable sortable'
r=requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser') 
seoul_ds=soup.find('table', class_="wikitable sortable")
df=pd.read_html(str(seoul_ds))
# convert list to dataframe
df=pd.DataFrame(df[0])
print(df.head())
```

```python
df[['Name','Name2']] = df.Name.str.split('\(|\)', expand=True).iloc[:,[0,1]]
df[['Name2','Name3']] = df.Name2.str.split(';', expand=True).iloc[:,[0,1]]
df=df.drop(columns=['Name3'])
df = df.rename(columns={'Name':'District','Name2':'Gu'})
```

```python
address = 'Seoul'

geolocator = Nominatim(user_agent="foursquare_agent")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(latitude, longitude)
```

```python
long =[]
lat = []

for i in df['District']:
    address = i 
    geolocator = Nominatim(user_agent="foursquare_agent")
    location = geolocator.geocode(address)
    latitude = location.latitude
    longitude = location.longitude
    long.append(longitude)
    lat.append(latitude)

df['Latitude']=lat
df['Longitude']=long
```

```python
# instantiate a feature group for the incidents in the dataframe
districts = folium.map.FeatureGroup()

for lat, lng, in zip(df_uni.Latitude, df_uni.Longitude):
    districts.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=False
        )
    )

# add pop-up text to each marker on the map
latitudes = list(df_uni.Latitude)
longitudes = list(df_uni.Longitude)
labels = list(df_uni.District)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(map_coffee)    
    
# add incidents to map
map_coffee.add_child(districts)
```
